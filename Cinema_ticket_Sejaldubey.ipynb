# Checking for stationarity

```python
df.set_index('date', inplace=True)

# Aggregate total sales by date
total_sales = df['total_sales'].resample('M').sum()
```

```python
total_sales.head()
```

```python
# Plot the time series
plt.figure(figsize=(10, 6))
plt.plot(total_sales, label='Total Sales', color='blue')
plt.title('Total Sales Over Time')
plt.xlabel('Date')
plt.ylabel('Total Sales')
plt.legend()
plt.show()
```

```python
from statsmodels.tsa.stattools import adfuller

# Perform the ADF test
adf_result = adfuller(total_sales.dropna())

print('ADF Statistic:', adf_result[0])
print('p-value:', adf_result[1])
print('Critical Values:')
for key, value in adf_result[4].items():
    print(f'   {key}: {value}')

if adf_result[1] > 0.05:
    print("The time series is non-stationary.")
else:
    print("The time series is stationary.")
```

make it stationary by differencing time series

```python
# Differencing the time series
total_sales_diff = total_sales.diff().dropna()

# Plot the differenced series
plt.figure(figsize=(10, 6))
plt.plot(total_sales_diff, label='Differenced Total Sales')
plt.title('Differenced Total Sales Over Time')
plt.xlabel('Date')
plt.ylabel('Differenced Total Sales')
plt.legend()
plt.show()

# Perform the ADF test on the differenced series
adf_result_diff = adfuller(total_sales_diff)
print('ADF Statistic:', adf_result_diff[0])
print('p-value:', adf_result_diff[1])
print('Critical Values:')
for key, value in adf_result_diff[4].items():
    print(f'   {key}: {value}')

if adf_result_diff[1] > 0.05:
    print("The differenced time series is non-stationary.")
else:
    print("The differenced time series is stationary.")
```

# ARIMA modeling

```python
from statsmodels.tsa.stattools import adfuller
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
from statsmodels.tsa.arima.model import ARIMA
```

```python
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
import matplotlib.pyplot as plt

# Plot ACF and PACF to determine p and q
fig, axes = plt.subplots(1, 2, figsize=(16, 6))

# ACF plot
plot_acf(total_sales, ax=axes[0])
axes[0].set_title('ACF Plot')

# PACF plot
plot_pacf(total_sales, ax=axes[1])
axes[1].set_title('PACF Plot')

plt.show()
```

```python
from statsmodels.tsa.arima.model import ARIMA

# Fit the ARIMA model
p = 1  # replace with value from PACF plot
d = 0  # as the series is already stationary
q = 1  # replace with value from ACF plot

model = ARIMA(total_sales, order=(p, d, q))
model_fit = model.fit()

# Print model summary
print(model_fit.summary())
```

```python
# Forecast future sales
forecast_steps = 12  # Number of months to forecast
forecast = model_fit.get_forecast(steps=forecast_steps)
forecast_index = pd.date_range(start=total_sales.index[-1], periods=forecast_steps+1, freq='M')[1:]

# Extract forecast mean and confidence intervals
forecast_mean = forecast.predicted_mean
forecast_ci = forecast.conf_int()

# Plot the forecast
plt.figure(figsize=(12, 6))
plt.plot(total_sales, label='Observed', color='blue')
plt.plot(forecast_index, forecast_mean, label='Forecast', color='orange')
plt.fill_between(forecast_index, forecast_ci.iloc[:, 0], forecast_ci.iloc[:, 1], color='k', alpha=0.1)
plt.title('Sales Forecast')
plt.xlabel('Date')
plt.ylabel('Total Sales')
plt.legend()
plt.show()
```

```python
from sklearn.metrics import mean_absolute_error

# Create a train-test split
train = total_sales[:int(0.8*len(total_sales))]
test = total_sales[int(0.8*len(total_sales)):]

# Fit the model on the training data
model = ARIMA(train, order=(p, d, q))
model_fit = model.fit()

# Make predictions on the test data
predictions = model_fit.forecast(steps=len(test))

# Calculate MAE
mae = mean_absolute_error(test, predictions)
print(f'Mean Absolute Error: {mae}')

# Plot the test and predicted values
plt.figure(figsize=(12, 6))
plt.plot(train, label='Train', color='blue')
plt.plot(test, label='Test', color='green')
plt.plot(test.index, predictions, label='Predicted', color='orange')
plt.title('ARIMA Model - Train, Test, and Predicted')
plt.xlabel('Date')
plt.ylabel('Total Sales')
plt.legend()
plt.show()
```
